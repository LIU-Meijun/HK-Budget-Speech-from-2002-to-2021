```python
import pandas as pd
```

```python
cos = pd.read_csv('../data/cosine_simi.csv', index_col=0)
```

```python
columns = cos.columns
values = []
indexs = []
cols = []
for index, row in cos.iterrows():
    for col in columns:
        value = row[col]
        values.append(value)

for index in range(1,941):
    for i in range(1,941):
        indexs.append(index)
     
    
for col in range(1, 941):
    for i in range(1, 941):
        cols.append(i)        

data = {'index' : indexs,
            'col': cols,
            'value': values}

df = pd.DataFrame(data)
df
        
```

```python
df.sort_values(by=['value'],ascending=[False],inplace=True)
df
```

```python
df1 = df.drop(df[df['index'] == df['col']].index)
df1
```

```python
df1.to_csv("../data/cosine_simi_sort.csv",index=False,sep=',')
```
